## Importing Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
warnings.filterwarnings("ignore")
```


## Importing  the  Dataset

```python
df = pd.read_csv('globalterrorismdb_0718dist.csv')
```

##  Data Preprocessing

```python
# Renaming the columns for better understanding
df.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'State',
                       'region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target','nkill':'Killed',
                       'nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type',
                       'weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)
```

```python
# Drop out the irrelevant columns & considering important data for further processing

df=df[['Year','Month','Day','Country','State','Region','city','latitude','longitude','AttackType','Killed',
               'Wounded','Target','Summary','Group','Target_type','Weapon_type','Motive']]
```

```python
df.head()
```

## Exploratory Data Analysis

<br>

## Hot zone of Terrorism:

## Top Countries Affected by Terrorist Attacks

```python
df['Country'].value_counts().head()
```

```python
# Visualising the countries affected by terrorist attacks
df['Country'].value_counts()
plt.figure(figsize=(10,5))
sns.barplot(x=df['Country'].value_counts()[:15].index,y=df['Country'].value_counts().values[:15], palette = 'rocket')
plt.title('Countries that got attacked the most')
plt.xticks(rotation=90)
plt.xlabel('Countries')
plt.ylabel('Attacks')
plt.show()
```

### Observation:
* Iraq is the most affected country by terrorist attacks followed by Pakistan and Afghanistan

## Top Regions Affected by Terrorist Attacks

```python
df['Region'].value_counts().head()
```

```python
## Visualising the regions affected by terrorist attacks
df['Region'].value_counts()
plt.figure(figsize=(10,5))
sns.barplot(x=df['Region'].value_counts()[:14].index,y=df['Region'].value_counts().values[:14], palette = 'deep')
plt.title('Regions that  got attacked the most')
plt.xticks(rotation=90)
plt.xlabel('Regions')
plt.ylabel('Attacks')
plt.show()
```

### Observation :
* Middle East & North Africa is the most affected region by terrorist attacks followed by South Asia and South America

## Top Cities Affected by Terrorist Attacks

```python
df['city'].value_counts().head()
```

```python
## Visualising the cities that are most affected by terrorist attacks
df['city'].value_counts()
plt.figure(figsize=(10,5))
sns.barplot(x=df['city'].value_counts()[:15].index,y=df['Country'].value_counts().values[:15])
plt.title('Cities that are  attacked the most')
plt.xticks(rotation=90)
plt.xlabel('Cities')
plt.ylabel('Attacks')
plt.show()
```

### Observation:
* Baghdad is the most affected city by terrorist attacks followed by Karachi and Lima

<br>

## Other Important visualizations:

## Terrorist Attacks Over the Years

```python
df['Year'].value_counts()
plt.figure(figsize=(10,5))
sns.barplot(x=df['Year'].value_counts()[:15].index,y=df['Year'].value_counts().values[:15], palette = 'colorblind')
plt.title('Attacks per Year')
plt.xticks(rotation=90)
plt.xlabel('Years')
plt.ylabel('Attacks')
plt.show()
```

### Observation:  
* 1. There has been a surge in the number of terror attacks from the year 2012 onwards.
* 2. The year has witnessed the highest number of terror attacks reaching 16000 and beyond.
* 3. The amount of terror attacks has been relatively low in the years from 1988 - 2011

## Type of Attacks by Terrorists

```python
plt.figure(figsize=(15,8))
sns.countplot(x=df['AttackType'],order=df['AttackType'].value_counts().index,palette='Blues_d')
plt.xticks(rotation=80)
plt.xlabel('Attack Type')
plt.ylabel('Number of Attacks')
plt.show()
```

### Observation:
* The most preferred terrorist attack is bombing/explosion followed by armed assault and assasinations

## Groups hosting terror attacks 

```python
# Visualising the groups hosting terrorist attacks using a bar graph
df['Group'].value_counts().to_frame().drop('Unknown').head(10).plot(kind='bar',color='cyan',figsize=(20,10))
plt.xlabel("Terrorist groups",fontsize=20)
plt.ylabel("No. of Attacks",fontsize=20)
plt.show()
```

### Observation:
* Taliban group has hosted the highest number of attacks followed by Islamic State of Iraq and the Levant (ISIL)

## Properties Affected by Terrorist Attacks

```python
plt.subplots(figsize=(20,10))
sns.countplot(df["Target_type"],order=df['Target_type'].value_counts().index,palette="gist_heat",edgecolor=sns.color_palette("mako"));
plt.xticks(rotation=90)
plt.xlabel("Attacktype",fontsize=20)
plt.ylabel("count",fontsize=20)
plt.show()
```

### Observation:
* Private Citizens & Property is the most affected property followed by military & police

## Weapons Used for Terrorist Attacks

```python
df['Weapon_type'].value_counts().head() 
df['Weapon_type'].value_counts().head().plot(kind = 'pie', figsize = (18,8), fontsize=10, autopct= '%3.1f%%',shadow=True, startangle=135, legend=True, )
plt.ylabel("Types of weapons used ", size = 10)
plt.xlabel("Attacks Occured", size = 10)
plt.xticks(size = 15)
plt.yticks(size = 15)
```

### Observation: 
* We observe that explosives are the most widely used weapons during the terrorist attacks followed by firearms and unknown sources.
